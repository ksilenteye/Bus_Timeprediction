```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load and preprocess the data
df = pd.read_csv('Trip Data for ML Model.csv')
```

```python
df['Operated_date'] = pd.to_datetime(df['Operated_date'], format='%d-%m-%Y')
df['Sch_Trip_Start'] = pd.to_datetime(df['Sch_Trip_Start'], format='%d-%m-%Y %H:%M')
df['Sch_Trip_End'] = pd.to_datetime(df['Sch_Trip_End'], format='%d-%m-%Y %H:%M')
df['Actual_Trip_Start'] = pd.to_datetime(df['Actual_Trip_Start'], format='%d-%m-%Y %H:%M')
df['Actual_Trip_End'] = pd.to_datetime(df['Actual_Trip_End'], format='%d-%m-%Y %H:%M')

# Handling missing values (if any)
missing_values = df.isnull().sum()

# Feature Engineering
df['Route_Name'] = df['Route_Name'].astype('category').cat.codes
df['Operated_Day'] = df['Operated_Day'].astype('category').cat.codes
df['Sch_Trip_Start_Hour'] = df['Sch_Trip_Start'].dt.hour
df['Sch_Trip_Start_Minute'] = df['Sch_Trip_Start'].dt.minute
df['Sch_Trip_Start_Day'] = df['Sch_Trip_Start'].dt.day
df['Sch_Trip_Start_Month'] = df['Sch_Trip_Start'].dt.month
df['Sch_Trip_DayOfWeek'] = df['Sch_Trip_Start'].dt.dayofweek  # Monday=0, Sunday=6
df['Sch_Trip_Weekend'] = df['Sch_Trip_DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)
df['Trip_Duration_Minutes'] = (df['Actual_Trip_End'] - df['Sch_Trip_Start']).dt.total_seconds() / 60.0

# Add more interaction features
df['Route_Minute_Interaction'] = df['Route_Name'] * df['Sch_Trip_Start_Minute']
df['Hour_Minute_Interaction'] = df['Sch_Trip_Start_Hour'] * df['Sch_Trip_Start_Minute']

# Use a smaller sample of data to avoid memory issues
df = df.sample(frac=0.3, random_state=42)

# Define features and target
X = df[['Route_Name', 'Operated_Day', 'Sch_Trip_Start_Hour', 'Sch_Trip_Start_Minute', 
        'Sch_Trip_Start_Day', 'Sch_Trip_Start_Month', 'Sch_Km', 
        'Sch_Trip_DayOfWeek', 'Sch_Trip_Weekend']]
y = df['Trip_Duration_Minutes']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = GradientBoostingRegressor()
```

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, 
                           cv=5, n_jobs=-1, verbose=2)
```

```python
# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Best model from GridSearchCV
best_model = grid_search.best_estimator_

# Evaluate the model
y_pred = best_model.predict(X_test)
```

```python
# Calculate the error metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
# Feature importance
importances = best_model.feature_importances_
indices = np.argsort(importances)[::-1]

# Plotting feature importances
plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.bar(range(X_train.shape[1]), importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), X.columns[indices], rotation=90)
plt.tight_layout()
plt.show()
```
